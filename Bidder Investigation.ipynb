```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from collections import Counter

pd.options.display.mpl_style = 'default'
```

```python
bids = pd.read_csv('bids.csv')
training_data = pd.read_csv('train.csv', index_col='bidder_id')
```

```python
is_bot_training = training_data['outcome'].to_dict()
bids['is_bot'] = bids.bidder_id.apply(lambda x: is_bot_training.get(x, 2))
```

```python
def codify_column(data_frame, column):
    codes_for_column = data_frame[column].unique()
    codes_for_column = dict(zip(codes_for_column, range(len(codes_for_column))))

    return bids[column].apply(lambda x: codes_for_column[x])
```

```python
bids['bidder_id_code'] = codify_column(bids, 'bidder_id')
bids['merchandise_code'] = codify_column(bids, 'merchandise')
bids['device_code'] = codify_column(bids, 'device')
bids['country_code'] = codify_column(bids, 'country')
bids['ip_code'] = codify_column(bids, 'ip')
bids['url_code'] = codify_column(bids, 'url')
bids['auction_code'] = codify_column(bids, 'auction')
```

```python
bidders = bids.groupby('bidder_id')
bidders_iter = iter(bidders)
```

```python
def pie_plot(labels, slices, ax=None):
    cmap = plt.cm.Pastel2
    colors = cmap(np.linspace(0., 1., len(slices)))

    if ax is None:
        plt.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=None, labeldistance=1.1)
    else:
        ax.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=None, labeldistance=1.1)
    plt.axis('equal')

def pie_plot_with_counter(countable, ax=None):
    labels = list(Counter(countable).keys())
    slices = list(Counter(countable).values())
    pie_plot(labels, slices, ax)
```

```python
def plot_all_pie_plots(this_auction):
    fig, ax = plt.subplots(2, 3, figsize= (15, 12))

    labels=['unlabeled', 'bots', 'human']
    null_count = this_auction.is_bot.apply(lambda x: x == 2).sum()
    ones_count = this_auction.is_bot.apply(lambda x: x == 1).sum()
    zeros_count = this_auction.is_bot.apply(lambda x: x == 0).sum()
    slices = [null_count, ones_count, zeros_count]

    pie_plot(labels, slices, ax[0][0])

    pie_plot_with_counter(this_auction.country, ax[0][1])
    pie_plot_with_counter(this_auction.device, ax[1][0])
    pie_plot_with_counter(this_auction.merchandise, ax[1][1])
    pie_plot_with_counter(this_auction.url, ax[0][2])
    
    ax[0][0].set_title("Is Bot")
    ax[0][1].set_title("Country") 
    ax[0][2].set_title("Url") 
    ax[1][0].set_title("Device") 
    ax[1][1].set_title("Merchandise") 

    ax[-1, -1].axis('off')
```

```python
def print_combined_plots(this_auction, ax=None):
    print('This auction: "{}"'.format(this_auction.auction.tolist()[0]))
    print('Number of Bids: {}'.format(this_auction.shape[0]))
    print('Number of Bidders: {}'.format(this_auction.bidder_id.unique().size))
    print('Winner is_bot = {}'.format(this_auction.sort('time').tail(1)['is_bot'].tolist()[0]))
    print()
    if ax is not None:
        time = this_auction.time.values
        is_bot = this_auction.is_bot.values
        ax.scatter(time, is_bot, c=is_bot, label=is_bot, cmap=plt.cm.jet, alpha=0.4, s=200)
    else:
        this_auction.plot(kind='scatter', x='time', y='is_bot', c=this_auction['is_bot'], colormap=plt.cm.jet, alpha=0.4,
                  s=200, figsize=(20, 2))
    #plot_all_pie_plots(this_auction, ax)
    plt.legend()
```

```python
this_bidder = next(bidders_iter)[1]
auctions_for_this_bidder = this_bidder.auction.unique().tolist()
bids_for_this_bidders_auctions = bids[bids.auction.isin(auctions_for_this_bidder)]

print('This bidder is bot = {}'.format(this_bidder.head(1).is_bot.values[0]))
print()
is_this_bidder = bids_for_this_bidders_auctions['bidder_id'] == this_bidder.bidder_id.values[0]
bids_for_this_bidders_auctions.loc[is_this_bidder, 'is_bot'] = 3

number_of_auctions = len(auctions_for_this_bidder)

fig, ax = plt.subplots(number_of_auctions, 1)
fig.set_size_inches(13, number_of_auctions*3)

for i in range(number_of_auctions):
    these_bids = bids_for_this_bidders_auctions.query('auction == "{}"'.format(auctions_for_this_bidder[i]))
    
    if number_of_auctions > 1:
        print_combined_plots(these_bids, ax[i])
    else:
        print_combined_plots(these_bids, ax)
```
