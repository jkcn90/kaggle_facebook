```python
def score_occurences(bids_grouped_by_auctions, threshold=0.1):
    number_of_auctions = len(bids_grouped_by_auctions)
    occurances = [x for sublist in bids_grouped_by_auctions for x in sublist]
    
    unique_occurances_per_auction = [list(set(x)) for x in bids_grouped_by_auctions]
    potential_targets = Counter([x for sublist in unique_occurances_per_auction for x in sublist])
    targets = [value for value, count in potential_targets.items() if count > number_of_auctions * threshold]
    targets_hit = [x for x in occurances if x in targets]
    number_of_hits = len(targets_hit)
    hit_rate = number_of_hits / len(occurances)
    return number_of_hits, hit_rate
```

```python
closeness_threshold = [1, 1e11, 5e1, 1e12]

def score_occurences_accross_auctions(these_bids, column_name='time', 
                                      closeness_threshold=closeness_threshold, 
                                      percentage_of_auctions_threshold=0.1):
    bids_grouped_by_auctions_per_threshold = these_bids[1].groupby(['auction']).apply(
        lambda x: [x[column_name].values // threshold for threshold in closeness_threshold]
        ).values
    bids_grouped_by_auctions_per_threshold = np.array(list(zip(*bids_grouped_by_auctions_per_threshold)))
    scores = [score_occurences(x, percentage_of_auctions_threshold) for x in bids_grouped_by_auctions_per_threshold]
    #this_score_occurences_function = partial(score_occurences, threshold=percentage_of_auctions_threshold)
    #p = Pool(4)
    #scores = p.map(this_score_occurences_function, bids_grouped_by_auctions_per_threshold)
    return scores
```

```python
def get_occurance_features(column_name, closeness_threshold, percentage_of_auctions_threshold):
    this_score_occurences_accross_auctions_function = partial(score_occurences_accross_auctions,
                                                              column_name=column_name,
                                                              closeness_threshold=closeness_threshold,
                                                              percentage_of_auctions_threshold=percentage_of_auctions_threshold)
    
    p = Pool(4)
    occurance_features = p.map(this_score_occurences_accross_auctions_function, iter(bids_grouped_by_bidder_id))
    
    occurance_features = np.array(occurance_features)
    occurance_features = np.array([x.flatten().tolist() for x in occurance_features])
    occurance_features_train = occurance_features[train_index]
    occurance_features_test = occurance_features[test_index]
    p.close()
    return (occurance_features, occurance_features_train, occurance_features_test)
```

```python
def extract_categorical_features(records, threshold=1):    
    print('Applying One Hot Encoding')
    vec = feature_extraction.DictVectorizer()
    one_hot_encoded = vec.fit_transform(records)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded.shape))

    print('Dropping features with low variance')
    threshold = threshold * (1 - threshold)
    print('Threshold = {}'.format(threshold))
    selector = feature_selection.VarianceThreshold(threshold=threshold) 
    one_hot_encoded_and_thresholded = selector.fit_transform(one_hot_encoded)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_thresholded.shape))
    
    print('Running through PCA')
    pca = decomposition.PCA()
    one_hot_encoded_and_pca_variance_and_thresholded = pca.fit_transform(one_hot_encoded_and_thresholded.toarray())
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_pca_variance_and_thresholded.shape))
    
    print('Removing mean and scaling variance')
    features = preprocessing.StandardScaler().fit_transform(one_hot_encoded_and_pca_variance_and_thresholded)
    return features
```

```python
def print_accuracy_auc_scores(clf, features, labels, cv=5, n_jobs=-1):
    scores = cross_validation.cross_val_score(clf, features, labels, scoring='roc_auc', cv=cv, n_jobs=n_jobs)
    print('auc scores: {}'.format(scores))

    scores = cross_validation.cross_val_score(clf, features, labels, cv=cv, n_jobs=n_jobs)
    print('accuracy scores: {}'.format(scores))
```
