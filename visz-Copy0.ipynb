```python
bids = raw_bids.copy()

bidder_id_outcome_map = dict(raw_training_data[['bidder_id', 'outcome']].values)
bids['outcome'] = bids['bidder_id'].apply(lambda x: bidder_id_outcome_map.get(x, -1))

bids['price'] = bids.groupby(['auction'])[['time']].transform(lambda x: np.array(x).argsort().argsort() + 1)
bids['winning_bid'] = bids.groupby(['auction'])['time'].apply(lambda x: x == max(x))

bidder_id_codes_map = bids.bidder_id.unique()
bidder_id_codes_map = {this_id: i for i, this_id in enumerate(bidder_id_codes_map)}

bids.loc[:,'bidder_id_code'] = bids['bidder_id'].apply(lambda x: bidder_id_codes_map[x])
bids['time_since_last_bid'] = bids.groupby(['auction'])[['time']].transform(lambda x: np.ediff1d(x, to_begin=0))
```

```python
time_group_0_min, time_group_0_max = 9631916842105263, 9645558894736842
time_group_1_min, time_group_1_max = 9695580000000000, 9709222052631578
time_group_2_min, time_group_2_max = 9759243157894736, 9772885210526315

bids['time_group'] = bids.time.apply(lambda x: (0 if x >= time_group_0_min and x <= time_group_0_max else
                                                1 if x >= time_group_1_min and x <= time_group_1_max else
                                                2 if x >= time_group_2_min and x <= time_group_2_max else
                                               -1))

bids['time_group_0'] = bids.time_group.apply(lambda x: 1 if x == 0 else 0)
bids['time_group_1'] = bids.time_group.apply(lambda x: 1 if x == 1 else 0)
bids['time_group_2'] = bids.time_group.apply(lambda x: 1 if x == 2 else 0)
```

# Feature Discovery

```python
def score_occurences(bids_grouped_by_auctions, threshold=0.1):
    number_of_auctions = len(bids_grouped_by_auctions)
    occurances = [x for sublist in bids_grouped_by_auctions for x in sublist]
    
    unique_occurances_per_auction = [list(set(x)) for x in bids_grouped_by_auctions]
    potential_targets = Counter([x for sublist in unique_occurances_per_auction for x in sublist])
    targets = [value for value, count in potential_targets.items() if count > number_of_auctions * threshold]
    targets_hit = [x for x in occurances if x in targets]
    number_of_hits = len(targets_hit)
    hit_rate = number_of_hits / len(occurances)
    return number_of_hits, hit_rate
```

```python
closeness_threshold = [1, 1e11, 5e1, 1e12]

def score_occurences_accross_auctions(these_bids, column_name='time', 
                                      closeness_threshold=closeness_threshold, 
                                      percentage_of_auctions_threshold=0.1):
    bids_grouped_by_auctions_per_threshold = these_bids.groupby('auction').apply(
        lambda x: [x[column_name].values // threshold for threshold in closeness_threshold]
        ).values
    bids_grouped_by_auctions_per_threshold = np.array(list(zip(*bids_grouped_by_auctions_per_threshold)))
    scores = [score_occurences(x, percentage_of_auctions_threshold) for x in bids_grouped_by_auctions_per_threshold]
    #this_score_occurences_function = partial(score_occurences, threshold=percentage_of_auctions_threshold)
    #p = Pool(4)
    #scores = p.map(this_score_occurences_function, bids_grouped_by_auctions_per_threshold)
    return scores
```

```python
def plot_test(grouped_test, column_name='time', threshold=1e12):
    times = grouped_test.groupby('auction').apply(lambda x: x[column_name].values // threshold).values.tolist()
    times_auction = np.array([[val, i] for i in range(len(times)) for val in times[i]])
    x = times_auction[:, 0]
    y = times_auction[:, 1]
    return (x, y)

def plot_vertical_test(grouped_test, column_name='time', threshold=1e12, secondary_threshold=2):
    times = grouped_test.groupby('auction').apply(lambda x: x[column_name].values // threshold).values.tolist()
    list_of_times = list(set([item for sublist in times for item in sublist]))
    overlap_count = grouped_test.groupby('auction').apply(lambda x: x[column_name].values // threshold).apply(
                        lambda x_list: [sum(x == time for x in x_list) for time in list_of_times])
    overlap_count = overlap_count.apply(lambda x: [1 if i > 0 else 0 for i in x])
    overlap_count = np.array(overlap_count.values.tolist())
    overlap_count = overlap_count.sum(axis=0)
    vertical_lines = list(zip(list_of_times, overlap_count))
    vertical_lines = [time for time, count in vertical_lines if count > secondary_threshold]
    return vertical_lines
```

```python
def threshold_test(_id, column_name='time', 
                   thresholds = [1, 1e11, 2e11, 3e11, 4e11, 5e11, 6e11, 7e11, 8e11, 9e11, 1e12, 12e12]):
    grouped_test = bids.query('bidder_id_code == {}'.format(_id))

    
    number_of_columns = 3
    number_of_rows = math.ceil(len(thresholds) / number_of_columns)
    if number_of_rows == 1:
        fig, ax_list = plt.subplots(1, len(thresholds), figsize=(len(thresholds) / 3 * 20, 5*len(thresholds)))
    else:
        fig, ax_list = plt.subplots(number_of_rows, number_of_columns, figsize=(20, 5*number_of_columns))

    for i, threshold in enumerate(thresholds):
        if number_of_rows == 1:
            if len(thresholds) == 1:
                ax = ax_list
            else:
                ax = ax_list[i%number_of_columns]
        else:
            ax = ax_list[i//number_of_columns, i%number_of_columns]
        x, y = plot_test(grouped_test, column_name, threshold=threshold)
        ax.scatter(x, y, s=200, alpha=0.3)
        ax.set_yticks([])
        match = score_occurences_accross_auctions(grouped_test, column_name, closeness_threshold=[threshold])
        vertical_lines = plot_vertical_test(grouped_test, column_name, threshold=threshold)
        ax.set_title('{}'.format(match))
        for x_vertical in vertical_lines:
            ax.axvline(x_vertical, color='k', linestyle='--')
```
