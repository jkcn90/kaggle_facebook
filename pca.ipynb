# Import raw data

```python
raw_bids = pd.read_csv('bids.csv')
raw_training_data = pd.read_csv('train.csv')
raw_test_data = pd.read_csv('test.csv')
```

```python
bids = raw_bids
bids['outcome'] = bids['bidder_id'].apply(lambda x: bidder_id_outcome_map.get(x, -1))
```

# Feature Extraction

## Categorical Features

```python
bids.columns
```

```python
def extract_categorical_features(df, return_labels = False):
    print('Dataframe shape: {}\n'.format(df.shape))
    columns = ['bidder_id', 'auction', 'merchandise', 'device', 'country', 'ip']

    print('Aggregating data for the following columns:\n{}\n'.format(columns))
    grouped_df = df[columns].groupby('bidder_id').aggregate(lambda x: ','.join(x))
    records = grouped_df.to_dict('records')
    
    print('Applying One Hot Encoding')
    vec = feature_extraction.DictVectorizer(separator=',')
    one_hot_encoded = vec.fit_transform(records)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded.shape))
    
    print('Running through PCA')
    pca = decomposition.PCA()
    one_hot_encoded_and_pca = pca.fit_transform(one_hot_encoded.toarray())
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_pca.shape))
    
    print('Dropping features with low variance')
    threshold = 0.99
    threshold = threshold * (1 - threshold)
    print('Threshold = {}'.format(threshold))
    selector = feature_selection.VarianceThreshold(threshold=threshold) 
    one_hot_encoded_and_pca_variance_and_thresholded = selector.fit_transform(one_hot_encoded_and_pca)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_pca_variance_and_thresholded.shape))
    
    features = one_hot_encoded_and_pca_variance_and_thresholded
    labels = grouped_df.reset_index()['bidder_id'].apply(lambda x: bidder_id_outcome_map.get(x, -1)).values
    indicies = grouped_df.index
    return features, labels, indicies
```

```python
features, labels, indicies = extract_categorical_features(bids)
```

```python
features = preprocessing.StandardScaler().fit_transform(features)
```

```python
train_index = np.where(labels != -1)
test_index = np.where(labels == -1)

features_train, labels_train = features[train_index], labels[train_index]
features_test = features[test_index]
print('Feature Vectors Training: {}'.format(features_train.shape))
print('Feature Vectors Testing: {}'.format(features_test.shape))
```

```python
print('Decision Tree')
clf = tree.DecisionTreeClassifier(class_weight='auto')
scores = cross_validation.cross_val_score(clf, features_train, labels_train, scoring='roc_auc', cv=5, n_jobs=-1)
print('auc scores: {}'.format(scores))

clf = tree.DecisionTreeClassifier(class_weight='auto')
scores = cross_validation.cross_val_score(clf, features_train, labels_train, cv=5, n_jobs=-1)
print('accuracy scores: {}'.format(scores))
```

```python
figure = plt.figure(figsize=(10, 5)) 
h = .02  # step size in the mesh
name = 'test'

X = features_train[:,:2]
y = labels_train

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=.4)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])

# Plot the training points
ax = plt.subplot(1, 2, 1)

ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
# and testing points
ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6)
ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_xticks(())
ax.set_yticks(())

clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)

ax = plt.subplot(1, 2, 2)
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
if hasattr(clf, "decision_function"):
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
else:
    print(np.c_[xx.ravel(), yy.ravel()])
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

# Put the result into a color plot
Z = Z.reshape(xx.shape)
ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

# Plot also the training points
ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
# and testing points
ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
           alpha=0.6)

ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(name)
ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
        size=15, horizontalalignment='right')
```
