```python
import pandas as pd
import numpy as np
import random
import scipy
from multiprocessing import Pool
from matplotlib.colors import ListedColormap
from collections import Counter

from sklearn import cross_validation
from sklearn import ensemble
from sklearn import tree
from sklearn import metrics
from sklearn import feature_extraction
from sklearn import feature_selection
from sklearn import grid_search
from sklearn import decomposition
from sklearn import svm
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.grid_search import GridSearchCV
```

# Import raw data

```python
raw_bids = pd.read_csv('bids.csv')
raw_training_data = pd.read_csv('train.csv')
raw_test_data = pd.read_csv('test.csv')
```

# Set up bids data

```python
training_bidder_ids = raw_training_data.bidder_id.values
test_bidder_ids = raw_test_data.bidder_id.values
```

```python
bids = raw_bids.copy()

bids['price'] = bids.groupby(['auction'])[['time']].transform(lambda x: np.array(x).argsort().argsort() + 1)
bids['won_auction'] = bids.groupby(['auction'])['time'].apply(lambda x: x == max(x))

bidder_id_outcome_map = dict(raw_training_data[['bidder_id', 'outcome']].values)
bids['outcome'] = bids['bidder_id'].apply(lambda x: bidder_id_outcome_map.get(x, -1))
```

```python
time_group_1_min = 9631916842105263
time_group_1_max = 9645558894736842

time_group_2_min = 9695580000000000
time_group_2_max = 9709222052631578

time_group_3_min = 9759243157894736
time_group_3_max = 9772885210526315

bids['time_group'] = bids.time.apply(lambda x: (0 if x >= time_group_1_min and x <= time_group_1_max else
                                                1 if x >= time_group_2_min and x <= time_group_2_max else
                                                2))

bids['time_group_1'] = bids.time.apply(lambda x: 1 if x >= time_group_1_min and x <= time_group_1_max else 0)
bids['time_group_2'] = bids.time.apply(lambda x: 1 if x >= time_group_2_min and x <= time_group_2_max else 0)
bids['time_group_3'] = bids.time.apply(lambda x: 1 if x >= time_group_3_min and x <= time_group_3_max else 0)

def get_array_difference(x):
    arr = np.array(x)
    arr.sort()
    return np.ediff1d(np.concatenate([arr[:1], arr]))
bids['time_since_last_bid'] = bids.groupby(['auction', 'time_group'])[['time']].transform(get_array_difference)
```

# Clean Data

```python
bids['country'] = bids['country'].fillna('none')
```

# Feature Extraction

## Time Based Features

```python
metadata_columns = ['won_auction', 'time_group_1', 'time_group_2', 'time_group_3']
time_metadata = bids.groupby('bidder_id')[metadata_columns].sum().values
number_of_times_starting_auction = bids.groupby('bidder_id')['time_since_last_bid'].apply(lambda x: sum(x == 0)).values
mean_time_since_last_bid = bids.groupby('bidder_id')['time_since_last_bid'].mean().values

features = np.column_stack([time_metadata, number_of_times_starting_auction, mean_time_since_last_bid])
names = np.array(metadata_columns + ['number_of_times_starting_auction', 'mean_time_since_last_bid'])
```

## Compare Models

```python
features = preprocessing.StandardScaler().fit_transform(features)

labels = bids.groupby('bidder_id').outcome.apply(lambda x: np.mean(x)).values
train_index = np.where(labels != -1)
test_index = np.where(labels == -1)

features_train = features[train_index]
features_test = features[test_index]
labels_train = labels[train_index]
```

```python
clf = tree.DecisionTreeClassifier(class_weight='auto')
clf.fit(features_train, labels_train)
feature_importances = clf.feature_importances_
threshold = 0.23

print('Feature Importance')
print('\n'.join(str(x[0]) + ': ' + str(x[1]) for x in zip(names, feature_importances)))
print()
features_over_threshold_index = feature_importances > threshold

print('Features over threshold of {}: {}'.format(threshold, sum(features_over_threshold_index)))
print('\n'.join(str(x[0]) + ': ' + str(x[1]) for x in 
                zip(names[feature_importances > 0.1], feature_importances[features_over_threshold_index])))
print()

features_thresholded = np.squeeze(features[:,np.where(features_over_threshold_index)])
features_train_thresholded = features_thresholded[train_index]
features_test_thresholded = features_thresholded[test_index]
```

```python
def get_match_features_for_bidder_id(bids_grouped_this_by_bidder_id):
    bids_grouped_this_by_bidder_id = bids_grouped_this_by_bidder_id[1]
    bids_grouped_by_auctions = bids_grouped_this_by_bidder_id.groupby('auction').apply(
    lambda x: x['time'].values // threshold).values.tolist()
    
    bids_grouped_by_auctions_sets = [set(x) for x in bids_grouped_by_auctions]
    bids_grouped_by_auctions_counters = [Counter(x) for x in bids_grouped_by_auctions]
    
    number_of_auctions = len(bids_grouped_by_auctions_sets)
    
    match_rate = []
    match_count = []
    for i in range(number_of_auctions):
        for j in range(i + 1, number_of_auctions):
            matches = bids_grouped_by_auctions_sets[i].intersection(bids_grouped_by_auctions_sets[j])
            matches_i = sum([bids_grouped_by_auctions_counters[i].get(k) for k in matches])
            matches_j = sum([bids_grouped_by_auctions_counters[j].get(k) for k in matches])
            matches = matches_i + matches_j
            total = sum(bids_grouped_by_auctions_counters[i].values()) + sum(bids_grouped_by_auctions_counters[j].values())

            match_rate.append(matches / total)
            match_count.append(matches)
    
    if match_rate:
        match_rate_mean = np.mean(match_rate)
        match_rate_mode = scipy.stats.mode(match_rate)[0][0]
        match_rate_min = min(match_rate)
        match_rate_max = max(match_rate)
        match_rate_std = np.std(match_rate)
        return [sum(match_count), match_rate_mean, match_rate_mode, match_rate_min, match_rate_max, match_rate_std]
    else:
        return [0, 0, 0, 0, 0, 0]
```

```python
bids_grouped_by_bidder_id = bids.groupby('bidder_id')

p = Pool(5)
match_features = p.map(get_match_features_for_bidder_id, iter(bids_grouped_by_bidder_id))
```

```python
match_features = np.array(match_features)
match_features_train = match_features[train_index]
match_features_test = match_features[test_index]
```

```python
combined_features = np.column_stack([features_thresholded, match_features])
combined_features_train = combined_features[train_index]
combined_features_test = combined_features[test_index]
```

```python
def extract_categorical_features(records, threshold=1):    
    print('Applying One Hot Encoding')
    vec = feature_extraction.DictVectorizer()
    one_hot_encoded = vec.fit_transform(records)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded.shape))

    print('Dropping features with low variance')
    threshold = threshold * (1 - threshold)
    print('Threshold = {}'.format(threshold))
    selector = feature_selection.VarianceThreshold(threshold=threshold) 
    one_hot_encoded_and_thresholded = selector.fit_transform(one_hot_encoded)
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_thresholded.shape))
    
    print('Running through PCA')
    pca = decomposition.PCA()
    one_hot_encoded_and_pca_variance_and_thresholded = pca.fit_transform(one_hot_encoded_and_thresholded.toarray())
    print('Size for Feature Vectors: {}\n'.format(one_hot_encoded_and_pca_variance_and_thresholded.shape))
    
    print('Removing mean and scaling variance')
    features = preprocessing.StandardScaler().fit_transform(one_hot_encoded_and_pca_variance_and_thresholded)
    return features
```

```python
grouped_categorical_features = bids.groupby('bidder_id').aggregate(lambda x: ','.join(x))
grouped_categorical_features = grouped_categorical_features.applymap(lambda x: Counter(x.split(',')))
print('Grouped bids size: {}'.format(grouped_categorical_features.shape))

columns = ['auction', 'merchandise', 'device', 'country', 'ip', 'url']
records = grouped_categorical_features[columns].sum(axis=1).values
```

```python
categorical_features = extract_categorical_features(records, 0.3)
```

```python
categorical_features_train = categorical_features[train_index]
categorical_features_test = categorical_features[test_index]
```

```python
super_combined_features = np.column_stack([combined_features, categorical_features])
super_combined_features_train = super_combined_features[train_index]
super_combined_features_test = super_combined_features[test_index]
```

```python
clf = tree.DecisionTreeClassifier(class_weight='auto')
clf.fit(super_combined_features_train, labels_train)
feature_importances = clf.feature_importances_
threshold = 0.022

print('Feature Importance')
print('\n'.join(str(x[0]) + ': ' + str(x[1]) for x in zip(names, feature_importances)))
print()
features_over_threshold_index = feature_importances > threshold

#print('Features over threshold of {}: {}'.format(threshold, sum(features_over_threshold_index)))
#print('\n'.join(str(x[0]) + ': ' + str(x[1]) for x in 
#                zip(names[feature_importances > 0.1], feature_importances[features_over_threshold_index])))
#print()

super_combined_features_thresholded = np.squeeze(super_combined_features[:,np.where(features_over_threshold_index)])
super_combined_features_train_thresholded = super_combined_features_thresholded[train_index]
super_combined_features_test_thresholded = super_combined_features_thresholded[test_index]
```

```python
k = -1
X_inputs = [(features, features_train), (features_thresholded, features_train_thresholded), 
            (match_features, match_features_train), (combined_features, combined_features_train),
            (categorical_features, categorical_features_train), 
            (super_combined_features_thresholded, super_combined_features_train_thresholded)]

X, X_train = X_inputs[k]
#pca = decomposition.KernelPCA(n_components= 4, kernel='rbf')
#X = pca.fit_transform(X)
plot_dimensions = X.shape[1]
fig_width = 3 * plot_dimensions
fig_width = min(fig_width, 20)

fig, ax_list = plt.subplots(plot_dimensions, plot_dimensions, figsize=(fig_width, fig_width))
fig.tight_layout(pad=0)

for i in range(plot_dimensions):
    for j in range(plot_dimensions):
        if i <= j:
            x = X_train[:,i]
            y = X_train[:,j]
            c = labels_train
        else:
            x = X[:,i]
            y = X[:,j]
            c = labels

        ax = ax_list[i, j]
        if i == j-float('Inf'):
            ax.text(0, 0.5, names[i], rotation=45, fontsize=100/fig_width, fontweight='bold')
        else:
            ax.scatter(x, y, s=200, alpha=0.3, c=c,)# cmap=plt.cm.Paired)
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
def print_accuracy_auc_scores(clf, features, labels, cv=5, n_jobs=-1):
    scores = cross_validation.cross_val_score(clf, features, labels, scoring='roc_auc', cv=cv, n_jobs=n_jobs)
    print('auc scores: {}'.format(scores))

    scores = cross_validation.cross_val_score(clf, features, labels, cv=cv, n_jobs=n_jobs)
    print('accuracy scores: {}'.format(scores))
```

```python
def print_accuracy_auc_scores_for_all_features(clf, clf_name='Classifier'):
    print(clf_name)
    print_accuracy_auc_scores(clf, features_train, labels_train)

    print('\n{} Thresholded'.format(clf_name))
    print_accuracy_auc_scores(clf, features_train_thresholded, labels_train)
    
    print('\n{} Match'.format(clf_name))
    print_accuracy_auc_scores(clf, match_features_train, labels_train)
    
    print('\n{} Combined'.format(clf_name))
    print_accuracy_auc_scores(clf, combined_features_train, labels_train)
    
    print('\n{} Super'.format(clf_name))
    print_accuracy_auc_scores(clf, super_combined_features_train_thresholded, labels_train)
```

```python
decision_tree_clf = tree.DecisionTreeClassifier(class_weight='auto')

param_grid = dict(criterion=['gini', 'entropy'],
                  class_weight=['auto', None],
                  min_samples_split=list(range(1,200)),
                  )
decision_tree_clf = GridSearchCV(estimator=decision_tree_clf, param_grid=param_grid, n_jobs=-1, scoring='roc_auc')
decision_tree_clf.fit(super_combined_features_train_thresholded, labels_train)
decision_tree_clf.best_estimator_
```

```python
clf = decision_tree_clf.best_estimator_
print_accuracy_auc_scores_for_all_features(clf, 'Decision Tree')
```

```python
estimator = random_forest_clf.best_estimator_
print_accuracy_auc_scores_for_all_features(estimator, 'Random Forest')
```

# Visualize clasifiers

```python
clfs = [decision_tree_clf.best_estimator_, random_forest_clf.best_estimator_, extra_trees_clf.best_estimator_]
clfs = [tree.DecisionTreeClassifier(), ensemble.RandomForestClassifier(), ensemble.ExtraTreesClassifier()]

visualize_classifiers(match_features_train[:,:2], labels_train, clfs, add_noise=False)
```

```python
def visualize_classifiers(X, y, clfs, add_noise=False, base_plot_size=10, number_cv=5, 
                          scatter_size=200, test_size = 0.4, mesh_step=0.2):
    number_of_classifiers = len(clfs)
    names = ['Raw Points' if x is None else x.__class__.__name__ for x in clfs]
    
    figure = plt.figure(figsize=(base_plot_size, base_plot_size * number_of_classifiers)) 


    # Split train/test and prepare mesh grid for the decision boundary
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=test_size)

    boundary_excess = 0.5 * (scatter_size / 20)
    x_min, x_max = X[:, 0].min() - boundary_excess, X[:, 0].max() + boundary_excess
    y_min, y_max = X[:, 1].min() - boundary_excess, X[:, 1].max() + boundary_excess
    xx, yy = np.meshgrid(np.arange(x_min, x_max, mesh_step),
                         np.arange(y_min, y_max, mesh_step))

    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])

    for i in range(number_of_classifiers):
        ax = plt.subplot(number_of_classifiers, 1, i+1)

        estimator = clfs[i]
        if estimator is not None:
            clf = estimator
            clf.fit(X_train, y_train)

            # Plot the decision boundary 
            if hasattr(clf, "decision_function"):
                Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
                prediction = clf.decision_function(X_test)
            else:
                Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
                prediction = clf.predict_proba(X_test)[:, 1]

            # Put the result into a color plot
            Z = Z.reshape(xx.shape)
            ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

            this_accuracy = clf.score(X_test, y_test)
            cv_accuracy = cross_validation.cross_val_score(clf, X, y, cv=number_cv, n_jobs=-1)
            cv_accuracy = ' '.join('{:.2f}'.format(x) for x in cv_accuracy)
            cv_accuracy = 'accuracy: {:.2f} ({})'.format(this_accuracy, cv_accuracy)

            this_auc_roc = metrics.roc_auc_score(y_test, prediction)
            cv_auc_roc = cross_validation.cross_val_score(clf, X, y, cv=number_cv, scoring='roc_auc', n_jobs=-1)
            cv_auc_roc = ' '.join('{:.2f}'.format(x) for x in cv_auc_roc)
            cv_auc_roc = 'auc roc: {:.2f} ({})'.format(this_auc_roc, cv_auc_roc)

            train_test_size = 'Size (train, test): ({}, {})'.format(X_train.shape[0], X_test.shape[0])

            true_human_index = np.where(y_test == 0)[0]
            true_robot_index = np.where(y_test > 0)[0]

            predicted_human_index = np.where(prediction == 0)[0]
            predicted_robot_index = np.where(prediction > 0)[0]

            number_of_predicted_humans = len(predicted_human_index)
            number_of_predicted_robots = len(predicted_robot_index)

            number_of_humans = len(true_human_index)
            number_of_robots = len(true_robot_index)

            number_of_humans_text = 'human bidders (predicted, actual): ({}, {})'.format(
                number_of_predicted_humans, number_of_humans)
            number_of_robots_text = 'robot bidders (predicted, actual): ({}, {})'.format(
                number_of_predicted_robots, number_of_robots)

            number_correctly_predicted_as_robot = len([x for x in predicted_robot_index if x in true_robot_index])
            number_incorrectly_predicted_as_robot = len([x for x in predicted_robot_index if x not in true_robot_index])
            number_missed_robot = len([x for x in true_robot_index if x not in predicted_robot_index])


            correctly_predicted_as_robots = 'correctly predicted as robot (TP): {} ({:.2f}%)'
            correctly_predicted_as_robots = correctly_predicted_as_robots.format(
                number_correctly_predicted_as_robot, 
                (100 * number_correctly_predicted_as_robot/number_of_predicted_robots
                 if number_of_predicted_robots > 0 else float('Inf')))

            incorrectly_predicted_as_robots = 'incorrectly predicted as robot (FP): {} ({:.2f}%)'
            incorrectly_predicted_as_robots = incorrectly_predicted_as_robots.format(
                number_incorrectly_predicted_as_robot, 
                (100 * number_incorrectly_predicted_as_robot/number_of_predicted_robots
                 if number_of_predicted_robots > 0 else float('Inf')))

            missed_robot_rate = 'robots missed (FN): {} ({:.2f}%)'
            missed_robot_rate = missed_robot_rate.format(
                number_missed_robot,
                (100 * number_missed_robot/number_of_robots if number_of_robots > 0 else float('Inf')))

            text_items = [cv_accuracy, cv_auc_roc, train_test_size,
                          number_of_humans_text, number_of_robots_text,
                          correctly_predicted_as_robots,
                          incorrectly_predicted_as_robots,
                          missed_robot_rate,
                          ]
            for j, text in enumerate(text_items):
                ax.text(1.1*xx.max(), yy.max() - 1*(j+1), text, size=15, horizontalalignment='left')

        # Plot the train/test data
        if add_noise==True:
            def add_noise_f(feat):
                return [x + scatter_size * 0.02 * random.random() * pow(-1, random.randint(0, 1)) for x in feat]
            x0, x1 = add_noise_f(X_train[:, 0]), add_noise_f(X_train[:, 1])
            x0_train, x1_train = add_noise_f(X_test[:, 0]), add_noise_f(X_test[:, 1])
        else:
            x0, x1 = X_train[:, 0], X_train[:, 1]
            x0_train, x1_train = X_test[:, 0], X_test[:, 1]

        ax.scatter(x0, x1, s=scatter_size, c=y_train, cmap=cm_bright)
        ax.scatter(x0_train, x1_train, s=scatter_size, c=y_test, cmap=cm_bright, alpha=0.6)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        ax.set_title(names[i])
```
