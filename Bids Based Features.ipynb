```python
def get_match_features_for_bidder_id(bids_grouped_this_by_bidder_id, column_name='time',
                                     thresholds=[1, 1e11, 2e11, 3e11, 4e11, 5e11, 6e11, 7e11, 8e11, 9e11, 1e12, 12e12]):
    bids_grouped_this_by_bidder_id = bids_grouped_this_by_bidder_id[1]
    match_features = []
    for threshold in thresholds:
        this_match_featues = get_match_features(bids_grouped_this_by_bidder_id, column_name=column_name, threshold=threshold)
        match_features.append(this_match_featues)
    return match_features
```

```python
def get_match_features(these_bids, column_name='time', threshold=1e12):
    bids_grouped_by_auctions = these_bids.groupby('auction').apply(
    lambda x: x[column_name].values // threshold).values.tolist()
    
    bids_grouped_by_auctions_sets = [set(x) for x in bids_grouped_by_auctions]
    bids_grouped_by_auctions_counters = [Counter(x) for x in bids_grouped_by_auctions]
    number_of_auctions = len(bids_grouped_by_auctions_sets)
    
    match_rate = []
    match_count = []
    for i in range(number_of_auctions):
        for j in range(i + 1, number_of_auctions):
            matches = bids_grouped_by_auctions_sets[i].intersection(bids_grouped_by_auctions_sets[j])
            matches_i = sum([bids_grouped_by_auctions_counters[i].get(k) for k in matches])
            matches_j = sum([bids_grouped_by_auctions_counters[j].get(k) for k in matches])
            matches = matches_i + matches_j
            total = sum(bids_grouped_by_auctions_counters[i].values()) + sum(bids_grouped_by_auctions_counters[j].values())
            match_rate.append(np.array(matches / total))
            match_count.append(matches)
    if match_rate:
        match_rate_mean = np.mean(match_rate)
        return [sum(match_count), match_rate_mean]
    else:
        return [1, 1]
```

```python
def print_accuracy_auc_scores(clf, features, labels, cv=5, n_jobs=-1):
    scores = cross_validation.cross_val_score(clf, features, labels, scoring='roc_auc', cv=cv, n_jobs=n_jobs)
    print('auc scores: {}'.format(scores))

    scores = cross_validation.cross_val_score(clf, features, labels, cv=cv, n_jobs=n_jobs)
    print('accuracy scores: {}'.format(scores))
```

```python
def print_accuracy_auc_scores_for_all_features(clf, clf_name='Classifier'):
    print(clf_name)
    print_accuracy_auc_scores(clf, features_train, labels_train)

    print('\n{} Match'.format(clf_name))
    print_accuracy_auc_scores(clf, match_features_train, labels_train)
    
    print('\n{} Match Time Since Last Bid'.format(clf_name))
    print_accuracy_auc_scores(clf, match_features_train_time_since_last_bid, labels_train)
    
    print('\n{} Combined'.format(clf_name))
    print_accuracy_auc_scores(clf, combined_features_train, labels_train)
    
    print('\n{} PCA Combined'.format(clf_name))
    print_accuracy_auc_scores(clf, pca_combined_features_train, labels_train)
```
